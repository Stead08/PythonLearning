```python
from sklearn import datasets
import pandas as pd 
dataset = datasets.load_boston()
#DataFrameに代入する
dataset2 = pd.DataFrame(dataset.data)
#列名を指定する
dataset2.columns = dataset.feature_names

#価格を取り出す
dataset3 = pd.DataFrame(dataset.target)
#列名を任意に指定する
dataset3.columns = ['target']
#二つのデータセットを縦に繋ぎ合わせる
dataset4 = pd.concat([dataset3, dataset2], axis=1)
#結果を出力
dataset4
```

```python
import pandas as pd
import numpy as np
import time
import urllib.request
#urllib.request をインポート
from urllib import request
#Beautiful soupをインポート
from bs4 import BeautifulSoup

#URLを指定する
url1 = 'https://hojyokin-portal.jp/subsidies/'

data4 = pd.DataFrame()

#listを指定する
list = np.arange(2392, 0, -1)
#リストを最新版から開始して、古い情報まで繰り返す
for j in range(len(list)):
    try:
        data3 = pd.DataFrame()
        url2 = list[j]

        url3 = url1+str(url2)
        print(url3)

        #URLを読み込む
        response = request.urlopen(url3)
        dfs = pd.read_html(url3)
        #DataFrameで情報を取得する
        data3 = data3.append(dfs)
        #取得したデータの行列を転置する
        data3 = data3.T

        soup = BeautifulSoup(response)
        #h1タグ(見出し）から補助金名を取得する
        hojyokin_namae = \
            soup.find_all('h1', class_='p-subsidy__heading')
        #divタグから補助金お概要を取得する
        hojyokin_shousai = soup.find_all('div', class_='p-subsidy__text')
        #内容を絞り込み、開業で分類する
        test2 = hojyokin_namae[0].text
        test3 = hojyokin_shousai[0].text

        #取り込んだデータにラベルをつけ
        #取り込んだデータにラベルをつけdata3に追記する
        data3['補助金名'] = test2
        data3['補助金概要'] = test3
        print(data3)
        #dataに取得した内容をdata4に放り込む
        data4 = data4.append(data3)
        #20秒作業を中断する
        time.sleep(20)
        response.close()

    except urllib.error.HTTPError:
        print("not available")
    else:
        pass

#データで取得した結果をCSVに保存する
label = "hojokin__20221215_b.csv"
data4.to_csv(label)


        
```

```python
#東京23区及び市町村の年ごとの創業比率について,Resas apiから取得する
#利用するパッケージをインポートする
import json
import urllib.request

import pprint
import pandas as pd

#api_key.jsonに指定したAPIに読み込む
#ファイルはUtf-8でエンコーディング
with open('api_key.json', mode = "r", encoding='UTF-8') as f:
    api_key = json.load(f)

#東京の市区町村コード情報をRESAS APIから取得する
url_base = 'https://opendata.resas-portal.go.jp/api/v1/cities'

#東京都の券番号は13のため。一意に指定する
p = {'prefCode': 13}
url = url_base + '?' + urllib.parse.urlenconde(p)
req = urllib.request.Request(url, headers=api_key)

with urllib.request.urlopen(req) as response:
    data = response.read()

d = json.loads(data.decode())

#東京の市区町村コードをPanda Dataframe形式に変換する
tokyo_list = pd.io.json.json_normalize(d['result'], meta=['bigCityFlag', 'cityCode', 'cityName', 'prefCode'])

#東京23区＋市区町村の年単位の創業比率をRESASAPIから取得する
url_base = 'https://opendata.resas-portal.go.jp/api/v1/municipality/fundation/perYear'

#空のDataFrameを用意する
s = pd.DataFrame()

#東京都のデータを取得する
for j in tokyo_list.cityCode:
    p = {'prefCode': 13, 'cityCode': j}
    url = url_base + '?' + urllib.parse.urlenconde(p)
    #print(url)
    req = urllib.request.Request(url, headers=api_key)

    with urllib.request.urlopen(req) as response:
        data = response.read()

    d = json.loads(data.decode())
    #pprint.pprint(d)

    #各市区町村ごとのデータをJSONからDataFrame方式に変換し、kに放り込む
    k = pd.io.json.json_normalize(d['result'], record_path='data', meta=['cityCode', 'cityName', 'prefCode', 'prefName'])

    #print(k)

    #kの情報をsに放り込む
    s = s.append(k)
    #これを市区町村分繰り返す

#東京都のデータをまとめて標準出力する
print(s)
```
